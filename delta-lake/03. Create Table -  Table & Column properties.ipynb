**1. Table properties**
- COMMENT - allow you to document the purpose of the table
- TBLPROPERTIES - used to specify table level metadata or configuration settings

```sql
CREATE OR REPLACE TABLE demo.delta_lake.companies
( company_name STRING,
 founded_date DATE,
 country STRING)
COMMENT "This table contains information about some of the successful tech companies"
TBLPROPERTIES(sensitivity = 'true', 'delta.enableDeletionVectors' = 'true')
```

```sql
DESC EXTENDED demo.delta_lake.companies;
```

**2. Column properties**
- NOT NULL - enforce data integrity and quality 
- COMMENT  -  document the purpose or context of individual columns in a table

```sql
CREATE OR REPLACE TABLE demo.delta_lake.companies
( company_name STRING  NOT NULL,
 founded_date DATE COMMENT 'The date the company was founded',
 country STRING)
COMMENT "This table contains information about some of the successful tech companies"
TBLPROPERTIES(sensitivity = 'true', 'delta.enableDeletionVectors' = 'true')
```

```sql
DESC EXTENDED demo.delta_lake.companies;
```

**3. Column properties**
-  Generated Identity column -  used to generate an identity for example a primary key value
- Generated computed columns -  Automatically calculate and store derived values based on other columns in the same table

**- Generated Identity columns**
GENERATED {ALWAYS | BY DEFAULT} AS IDENTITY [([START WITH start] [INCREMENT BY step]) ]

```sql
CREATE OR REPLACE TABLE demo.delta_lake.companies
( company_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY( INCREMENT BY 1 START WITH 1),
 company_name STRING  NOT NULL,
 founded_date DATE COMMENT 'The date the company was founded',
 country STRING)
COMMENT "This table contains information about some of the successful tech companies"
TBLPROPERTIES(sensitivity = 'true', 'delta.enableDeletionVectors' = 'true')
```

```sql
INSERT INTO demo.delta_lake.companies (company_name, founded_date, country)
VALUES ('Airbnb', '2008-08-08', 'USA'),
('Dropbox', '2007-07-03', 'USA'),
('Salesforce', '1999-02-06', 'USA'),
('Zoom', '2011-03-12', 'USA'),
('Stripe', '2010-10-26', 'USA'),
('Slack', '2013-03-28', 'USA')
```

```sql
SELECT * FROM demo.delta_lake.companies;
```

**Generated computed columns **

GENERATED ALWAYS AS (expr)

`expr` may be composed of literals, column identifires within the table, and deterministic, build in SQL functions or operators except:
- Aggregate functions
- Analytic window functions
- Ranking window functions
- Table valued genrated functions

Also `expr`  must not contain any subquery

```sql
CREATE OR REPLACE TABLE demo.delta_lake.companies
( company_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY( INCREMENT BY 1 START WITH 1),
 company_name STRING  NOT NULL,
 founded_date DATE COMMENT 'The date the company was founded',
 founded_year INT GENERATED ALWAYS AS (YEAR(founded_date)),
 country STRING)
COMMENT "This table contains information about some of the successful tech companies"
TBLPROPERTIES(sensitivity = 'true', 'delta.enableDeletionVectors' = 'true')
```

```sql
INSERT INTO demo.delta_lake.companies (company_name, founded_date, country)
VALUES ('Airbnb', '2008-08-08', 'USA'),
('Dropbox', '2007-07-03', 'USA'),
('Salesforce', '1999-02-06', 'USA'),
('Zoom', '2011-03-12', 'USA'),
('Stripe', '2010-10-26', 'USA'),
('Slack', '2013-03-28', 'USA')
```

```sql
SELECT * FROM demo.delta_lake.companies;
```
