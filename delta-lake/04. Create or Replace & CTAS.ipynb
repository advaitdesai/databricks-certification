**1. Diffence between Drop and Create  & Create or Replace statements**

Drop and Create  -  delete the history
Create or Replace -  maintain the history

```sql
DESC HISTORY demo.delta_lake.companies;
```

**2. CTAS statment**

```sql
CREATE OR REPLACE TABLE demo.delta_lake.companies_usa
AS

SELECT * FROM 
demo.delta_lake.companies
WHERE country = "USA";
```

```sql
SELECT * FROM demo.delta_lake.companies_usa;
```

```sql
DESC EXTENDED demo.delta_lake.companies_usa;
```

```sql
CREATE OR REPLACE TABLE demo.delta_lake.companies_usa
AS

SELECT 
 CAST(company_id AS INT) AS company_id,
 company_name,
 founded_date,
 founded_year,
 country
FROM 
demo.delta_lake.companies
WHERE country = "USA";
```

```sql
DESC EXTENDED demo.delta_lake.companies_usa;
```

```sql
ALTER TABLE demo.delta_lake.companies_usa
 ALTER COLUMN company_id SET NOT NULL, 
 founded_year COMMENT 'founded date';
```

```sql
DESC EXTENDED demo.delta_lake.companies_usa;
```
