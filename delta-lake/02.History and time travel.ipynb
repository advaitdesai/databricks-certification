2. Query data from a specific version

```sql
SELECT * FROM demo.delta_lake.companies VERSION AS OF 1;
```

3. Query data from a Specific Time

```sql
SELECT * FROM demo.delta_lake.companies TIMESTAMP AS OF '2026-07-17T21:03:19.000+00:00';
```

```sql
RESTORE TABLE  demo.delta_lake.companies TO VERSION AS OF 1
```
