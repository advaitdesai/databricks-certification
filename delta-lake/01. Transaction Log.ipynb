## **Understanding DeltaLake Transaction Log**

Understand the cloud storage structure behind delta lake tables

0. create demo catalog

```python
%sql

CREATE CATALOG IF NOT EXISTS  demo
    MANAGED LOCATION 'abfss://demo@deacoursestacc.dfs.core.windows.net/' ;
```

1. Create a new schema under the demo catalog for this section of the course (delta_lake)

```python
%sql
USE CATALOG demo;
CREATE SCHEMA IF NOT EXISTS demo.delta_lake
  MANAGED LOCATION "abfss://demo@deacoursestacc.dfs.core.windows.net/delta_lake"
```

2. Create a Delta lake table

```python
%sql
CREATE OR REPLACE TABLE demo.delta_lake.companies(
    company_name STRING,
    founded_date DATE,
    country STRING
    )
```

```python
%sql
DESC EXTENDED demo.delta_lake.companies
```
