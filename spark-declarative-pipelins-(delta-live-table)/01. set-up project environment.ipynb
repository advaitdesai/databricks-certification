**1. Create external location**

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dea_course_ext_stac_circuitbox
    URL 'abfss://circuitbox@deacoursestacc.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL dea_course_ext_sc)
```

```sql
%fs ls 'abfss://circuitbox@deacoursestacc.dfs.core.windows.net/'
```

**2. create catalog**

```sql
CREATE CATALOG IF NOT EXISTS  circuitbox
    MANAGED LOCATION 'abfss://circuitbox@deacoursestacc.dfs.core.windows.net/' ;
```

```sql
SHOW CATALOGS;
```

**3. create Schema**

```sql
CREATE SCHEMA IF NOT EXISTS circuitbox.landing
    MANAGED LOCATION 'abfss://circuitbox@deacoursestacc.dfs.core.windows.net/landing';

CREATE SCHEMA IF NOT EXISTS circuitbox.lakehouse
    MANAGED LOCATION 'abfss://circuitbox@deacoursestacc.dfs.core.windows.net/lakehouse';
```

```sql
USE CATALOG circuitbox;
SHOW SCHEMAS;
```

4. **Create Volume**

```sql
CREATE  EXTERNAL VOLUME IF NOT EXISTS circuitbox.landing.operational_data
    LOCATION 'abfss://circuitbox@deacoursestacc.dfs.core.windows.net/landing/operational_data'
```

```sql
%fs ls '/Volumes/circuitbox/landing/operational_data/'
```
