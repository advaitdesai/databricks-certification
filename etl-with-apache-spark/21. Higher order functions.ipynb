**Higher order functions**
- Higher order functions are functions that operate on complex data types such as arrays and maps
- They allow you to pass fuctions as arguments ( such as lambda expressions), Apply transformations and return arrys or maps
- They are extremely useful for manipulationg arrays without exploding them.


**Commonly used Higher order Array functions**

- TRANSFORM
- FILTER
- EXISTS
- AGGREGATE

**Syntax**

< function_name > (array_column, lambda_exoression)

lambda_expression: element -> expression

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW order_items
AS
SELECT * FROM
VALUES (1, array('smartphone','laptop','monitor')),
        (2, array('tablet','headphone','smartwatch')),
        (3,array('keyboard','mouse'))
AS orders(order_id, items);
```

```python
%sql
SELECT * FROM order_items;
```

1. Convert all the item names to UPPERCASE (TRANSFORM function)

```python
%sql
SELECT order_id,
    TRANSFORM(items, x -> UPPER(x)) AS upper_items
FROM order_items;
```

2.Filter only items that contains the string 'smart' (FILTER function)

```python
%sql
SELECT order_id,
    FILTER(items, x -> x LIKE '%smart%') AS smart_items
FROM order_items;
```

3. check to see whether the order includes any 'monitor' (EXISTS function)

```python
%sql
SELECT order_id,
    EXISTS(items, x -> x LIKE 'monitor') AS monitor_items
FROM order_items;
```

**Array with more than one object**

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW order_items
AS
SELECT * FROM
VALUES (1, array(named_struct('name','smartphone', 'price', 699),
            named_struct('name','laptop', 'price', 1199),
            named_struct('name','monitor', 'price', 399)
        )),
        (2, array(named_struct('name','tablet', 'price', 599),
            named_struct('name','headphone', 'price', 199),
            named_struct('name','smartwatch', 'price', 299)
        )),
        (3, array(named_struct('name','keyboard', 'price', 89),
            named_struct('name','mouse', 'price', 59)    
        ))
AS orders(order_id, items);
```

```python
%sql
SELECT * FROM order_items;
```

1.convert all the item names to be UPPERCASE and add 10% TAx to each item (Transform function)

```python
%sql
SELECT order_id,
    TRANSFORM(items, x -> named_struct(
        'name', UPPER(x.name),
        'price', ROUND(x.price * 1.10,2)

    )) AS upper_items
FROM order_items;
```

2. Calclulate the total order amount for each of the order (AGGREGARETE function)

```python
%sql
SELECT order_id,
    AGGREGATE(items, 0,(acc,x) ->  acc + x.price) AS total
    
FROM order_items;
```

# **MAP functions**

A map is collection of key-value pairs, like a dictionary
{'laptop': 1200, 'phone': 699}

**Commonly used Higher order Map Functions**

- TRANSFORM_VALUES
- TRANSFORM_KEYS
- MAP_FILTER

**SYNTAX**

< function name > (map_column, lambda_expression)
lambda_expression : (key, value) -> expression

```python
%sql
CREATE OR REPLACE TEMP VIEW order_item_prices 
AS
SELECT * FROM VALUES
    (1, map('smartphone', 699, 'laptop', 1199, 'monitor', 399)),
    (2, map('tablet', 599, 'headphone', 199, 'smartwatch', 299)),
    (3, map('keyboard', 89, 'mouse', 59))
AS orders(order_id, item_prices);
    
```

```python
%sql
SELECT * FROM order_item_prices;
```

1. Convert all the items to be uppercase (TRANSFORM_KEYS function)

```python
%sql
SELECT 
order_id,
TRANSFORM_KEYS(item_prices,(x,y) -> UPPER(x)) AS upper_item_prices
FROM order_item_prices;
```

`Apply 10% Tax to item prices (TRANSFORM_VALUES function)`

```python
%sql
SELECT 
order_id,
TRANSFORM_VALUES(item_prices,(x,y) -> ROUND(y * 1.10,2)) AS upper_item_prices
FROM order_item_prices;
```

Filter only items with price above $500 (MAP_FILTER function)

```python
%sql
SELECT 
order_id,
MAP_FILTER(item_prices, (x,y) -> (y > 500)) AS filtered_items
FROM order_item_prices;
```
