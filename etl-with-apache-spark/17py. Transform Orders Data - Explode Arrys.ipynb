```python
df = spark.table('gizmobox.silver.py_orders_fixed')
display(df)
```

1.Access elements from the JSON object

```python
from pyspark.sql import functions as F


df_json_elements =(
    df
    .select(F.col('json_values').order_id.alias('order_id'),
            F.col('json_values').customer_id.alias('customer_id'),
            F.col('json_values').order_status.alias('order_status'),
            F.col('json_values').payment_method.alias('payment_method'),
            F.col('json_values').total_amount.alias('total_amount'),
            F.col('json_values').transaction_timestamp.alias('transaction_timestamp'),
            F.col('json_values').order_date.alias('order_date'),
            F.col('json_values').items.alias('items')
            )
)

display(df_json_elements)
```

2. Deduplicate Array elements

Function [array_distinct](https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/api/pyspark.sql.functions.array_distinct.html?highlight=array_distinct)

```python
from pyspark.sql import functions as F


df_json_elements =(
    df
    .select(F.col('json_values').order_id.alias('order_id'),
            F.col('json_values').customer_id.alias('customer_id'),
            F.col('json_values').order_status.alias('order_status'),
            F.col('json_values').payment_method.alias('payment_method'),
            F.col('json_values').total_amount.alias('total_amount'),
            F.col('json_values').transaction_timestamp.alias('transaction_timestamp'),
            F.col('json_values').order_date.alias('order_date'),
            F.array_distinct(F.col('json_values').items).alias('items')
            )
)

display(df_json_elements)
```

3. Explode Arrays (create multiple rows for arry eleements)

Fucnction [explode](https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/api/pyspark.sql.functions.explode.html)

```python
from pyspark.sql import functions as F


df_json_elements =(
    df
    .select(F.col('json_values').order_id.alias('order_id'),
            F.col('json_values').customer_id.alias('customer_id'),
            F.col('json_values').order_status.alias('order_status'),
            F.col('json_values').payment_method.alias('payment_method'),
            F.col('json_values').total_amount.alias('total_amount'),
            F.col('json_values').transaction_timestamp.alias('transaction_timestamp'),
            F.col('json_values').order_date.alias('order_date'),
            F.explode(F.array_distinct(F.col('json_values').items)).alias('item')
            )
)

display(df_json_elements)
```

```python
df_order_json = (
    df_json_elements
    .select(
            "order_id",
    "customer_id",
    "order_status",
    "payment_method",
    "total_amount",
    "transaction_timestamp",
    "order_date",
    "item.item_id",
    "item.name",
    "item.price",
    "item.quantity",
    "item.category",
    "item.details.brand",
    "item.details.color"
    )
)

display(df_order_json)
```

4. Write transformed Data to silve Schema

```python
df_order_json.writeTo('gizmobox.silver.py_orders').createOrReplace()
```
