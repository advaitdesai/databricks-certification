```python
df_customers = spark.table('gizmobox.silver.py_customers')
display(df_customers)
```

```python
df_addresses = spark.table('gizmobox.silver.py_addresses')
display(df_addresses)
```

```python
df_customers_addresses = (
    df_customers
    .join(df_addresses, on='customer_id', how='inner')
    .select(
        'customer_id',
        'customer_name',
        'date_of_birth',
        'email',
        'member_since',
        'telephone',
        'shipping_address_line_1',
        'shipping_city',
        'shipping_state',
        'shipping_postcode',
        'billing_address_line_1',
        'billing_city',
        'billing_state',
        'billing_postcode'
    )

    )

display(df_customers_addresses)
```

2. Write data to gold schema

```python
df_customers_addresses.writeTo('gizmobox.gold.py_customer_address').createOrReplace()
```
