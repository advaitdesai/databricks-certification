```python
df = spark.table('gizmobox.silver.py_orders')
display(df)
```

**1. Total orders**

```python
from pyspark.sql import functions as F

df_monthly_orders = (
    df
    .withColumn('order_month', F.date_format('transaction_timestamp','yyyy-MM'))
    .groupBy('order_month','customer_id')
    .agg(
        F.countDistinct('order_id').alias('total_orders'),
        F.sum('quantity').alias('total_item_bought'),
        F.sum(F.col('quantity') * F.col('price')).alias('total_amount')
    )

    
)

display(df_monthly_orders)
```

**2. Write Data to gold schema**

```python
df_monthly_orders.writeTo('gizmobox.gold.py_order_summary_monthly').createOrReplace()
```
