1. Run SQL command using python - spark.sql() function

```python
df = spark.sql("select * from json.`/Volumes/gizmobox/landing/operational_data/customers`")
display(df)
```

```python
spark.sql('''CREATE OR REPLACE TEMP VIEW tv_customers
          AS
          select * from json.`/Volumes/gizmobox/landing/operational_data/customers`''')
```

```python
spark.sql("select * from tv_customers").display()
```
