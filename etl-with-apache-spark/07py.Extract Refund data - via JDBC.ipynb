1.Read Refund data Via JDBC

```python
df =  (
    spark.read.format('jdbc')
    .option('jdbcUrl', 'jdbc:postgresql://database.databricks.training:5432/training')
    .option('dbtable', 'refunds')
    .option('user', 'training')
    .option('password', 'training')
    .load()
)
```
