**3.rofile Data manually**

**3.1  COUNT function**

```python
%sql
SELECT COUNT(*), COUNT(customer_id), COUNT( EMAIL) FROM gizmobox.bronze.v_customers;
    
```

**3.2 COUNT_IF Function**

```python
%sql
SELECT COUNT(*), COUNT_IF(customer_id IS NULL), COUNT_IF( EMAIL IS NULL) FROM gizmobox.bronze.v_customers;
```

**3.3 WHERE Clause**

```python
%sql
SELECT COUNT(*) FROM gizmobox.bronze.v_customers WHERE customer_id IS NULL;
```

**3.4 DISTINCT KEYWORD**

```python
%sql
SELECT COUNT(*) AS total_records , COUNT(DISTINCT customer_id) FROM gizmobox.bronze.v_customers WHERE customer_id IS NOT NULL
```
