```python
df = spark.table('gizmobox.bronze.py_memberships')
display(df)
```

1. Extract customer_id from the filee path

```python
from pyspark.sql.functions import regexp_extract, col

df_extracted = (
    df.
    select(regexp_extract('path','([^/]+)(?=\\.[^.]+$)', 1).alias('customer_id'),
           col('content').alias('membership_card'))

)

display(df_extracted)
```

2. Write Transformed data to the silver schema

```python
df_extracted.writeTo('gizmobox.silver.py_memberships').createOrReplace()
```
