Create external table and read data from sql servere via JDBC
- You can create it on via JDBC only in hive metastore 
- for Unity catalog, need to use Lakehouse federation

```python
%sql
CREATE SCHEMA IF NOT EXISTS hive_metastore.bronze;

CREATE TABLE IF NOT EXISTS hive_metastore.bronze.refunds
USING JDBC
OPTIONS(
    URL "<DB JDBC CONNECTION STRING>",
    dbtable  'refunds',
    user '<DB USERNAME>',
    password '<DB PASSWORD>'


)
```
