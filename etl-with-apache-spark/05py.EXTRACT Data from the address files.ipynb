2. Use read_files function to read the files 

```python
df = (spark.read.format('csv')
      .option('header', 'true')
      .option('delimiter', '\t')
      .load('/Volumes/gizmobox/landing/operational_data/addresses')

)
display(df)
```

```python
df.writeTo('gizmobox.bronze.py_addresses').createOrReplace()
```
