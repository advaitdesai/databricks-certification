```python
payment_schema = 'payment_id INTEGER, order_id INTEGER, payment_timestamp TIMESTAMP,payment_status INTEGER, payment_method STRING'
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, TimestampType

py_payment_schema = StructType([
    StructField('payment_id', IntegerType()),
    StructField('order_id', IntegerType()),
    StructField('payment_timestamp', TimestampType()),
    StructField('payment_status', IntegerType()),
    StructField('payment_method', StringType())
])
```

```python
df = (
    spark.read.format('csv')
      .option('delimiter', ',')
      .schema(payment_schema)
      .load('abfss://gizmobox@deacoursestacc.dfs.core.windows.net/landing/external_data/payments')

)

display(df)
```

2. Create external table

```python
df.writeTo('gizmobox.bronze.py_payments').createOrReplace()
```
