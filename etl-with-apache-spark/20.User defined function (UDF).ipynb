**User Defined functions(UDF)**

Syntax
CREATE OR REPLACE FUNCTION <catalog namd.schem.function_name> (param data_type)

RETURNS return_datatype

RETURN expression;

Write a function to concatenate firstname and lastname to and output the fullname

```python
%sql
CREATE OR REPLACE FUNCTION gizmobox.default.get_fullname(fname STRING, lname STRING)
RETURNS STRING
RETURN CONCAT(initcap(fname), ' ', initcap(lname))
```

```python
%sql
SELECT  gizmobox.default.get_fullname('john', 'doe');
```

```python
%sql
DESC FUNCTION EXTENDED gizmobox.default.get_fullname;
```
